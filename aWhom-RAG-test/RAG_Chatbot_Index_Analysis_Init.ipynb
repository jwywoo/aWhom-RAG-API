```python
import os

from pinecone import Pinecone
pinecone_api_key = userdata.get('pinecone-KEY')
pc = Pinecone(api_key=pinecone_api_key)
```

```python
csv_path = os.path.join('/content/drive/MyDrive/Hackathon/CSV')
csv_files = [f for f in os.listdir(csv_path) if f.endswith('.csv')]
test_csv_path = os.path.join(csv_path, csv_files[0])
```

```python
def string_gen_from_queue(given_queue):
  doc_page_content = []
  for row in given_queue:
    doc_page_content.append(
        f"Index: {row['Index']} [SEP] Date : {row['Date']} [SEP] User : {row['User']} [SEP] Message : {row['Message']}"
    )
  return " ".join(doc_page_content)
```

```python
from langchain_openai import OpenAIEmbeddings
from langchain_pinecone import PineconeVectorStore

os.environ["OPENAI_API_KEY"] = userdata.get('openAI')
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
index = pc.Index(name="hackathon-awhom")
vector_store = PineconeVectorStore(index=index, embedding=embeddings)
```

```python
from uuid import uuid4
import pandas as pd
from langchain_core.documents import Document
df = pd.read_csv(test_csv_path)

documents = []
queue = []
for row in df.itertuples():
    row_dict = row._asdict()
    queue.append(row_dict)
    if (len(queue) == 5):
      doc_page_content = string_gen_from_queue(queue)
      documents.append(Document(
          page_content=doc_page_content
      ))
      queue.pop(0)

uuids = [str(uuid4()) for _ in range(len(documents))]
vector_store.add_documents(documents=documents, ids=uuids)
```
